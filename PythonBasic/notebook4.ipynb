```python
# tkinter
# tk Tcl
import tkinter
from tkinter import *
from tkinter import messagebox
import sqlite3

# db connection 
# con = sqlite3.connect('firstdb.db')

# get()
def get_data():
    con = sqlite3.connect('firstdb.db')
    n = ename.get()
    p = ephone.get()
    a = eaddress.get()
    con.cursor()
    con.execute("INSERT INTO contact(name, phone, address) values(?,?,?)",(n,p,a))
    con.commit()
    con.close()
    messagebox.showinfo('delete data','Data Save Successfully')
    retrive_data()

def delete_data():
    con = sqlite3.connect('firstdb.db')
    a = list.get(ACTIVE)
    b = a.split("-")[0]
    con.cursor()
    con.execute("delete from contact where name=?", (b,))
    con.commit()
    data = con.execute("select * from contact")
    print(data)
    con.close()
    # messagebox.showinfo('delete data','Data Delete Successfully')
    # messagebox.showerror('delete data','Data Delete Successfully')
    messagebox.showwarning('delete data','Data Delete Successfully')
    retrive_data()

def updat_data():
    n = ename.get()
    p = ephone.get()
    a = eaddress.get()
    active_data = list.get(ACTIVE)
    b = active_data.split("-")[0]
    con = sqlite3.connect('firstdb.db')
    con.cursor()
    con.execute("update contact set name=?, phone=?, address=? where name=?",(n,p,a,b))

    con.commit()
    con.close()
    retrive_data()
    ename.delete(0, END)
    ephone.delete(0, END)
    eaddress.delete(0, END)
    messagebox.showinfo('delete data','Data update Successfully')
        

def retrive_data():
    list.delete(0,END)
    con = sqlite3.connect('firstdb.db')
    con.cursor()
    data = con.execute("select * from contact")
    for i in data.fetchall():
        list.insert(END,f"{i[0]}-{i[1]}-{i[2]}")
    con.commit()
    con.close()
    print('load data')



        

frmae =tkinter.Tk()
frmae.title('this is title')
frmae.geometry('500x400')
tkinter.Label(text='name').grid(row=2, column=1)
tkinter.Label(text='phone').grid(row=3, column=1)
tkinter.Label(text='address').grid(row=4, column=1)
list = tkinter.Listbox(width=50)

tkinter.Button(text='Register', command=lambda:get_data()).grid(row=5, column=2)

tkinter.Button(text='Delete', command=lambda:delete_data()).grid(row=7, column=2)
tkinter.Button(text='Update', command=lambda:updat_data()).grid(row=7, column=3)

ename = tkinter.Entry()
ephone = tkinter.Entry()
eaddress = tkinter.Entry()

ename.grid(row=2, column=2)
ephone.grid(row=3, column=2)
eaddress.grid(row=4, column=2)
list.grid(row=6, column=2)

retrive_data()




frmae.mainloop()  #display window
```
